```python
import pandas as pd
```

```python
# load news accounts from class spreadsheet
def load_news_accounts(news_accounts_path):
    news_accounts = pd.read_csv(news_accounts_path)
    news_accounts.drop_duplicates(subset=['Username'], inplace=True)
    return news_accounts
```

```python
def load_pyktok_data(file):
    videos = pd.read_csv(file)
    return videos
```

```python
def get_news_content(video_df, news_accs):
    news_videos = video_df[video_df['author_username'].isin(news_accs['Username'])]
    news_videos.dropna(subset=["author_username"], inplace=True)
    return news_videos
```

```python
news = load_news_accounts('news_accounts.csv')

pyktok_files = ['results_AY.csv', 'results_CF.csv','results_AL.csv', 'results_JK.csv', 'results_MG.csv', 'results_chunk_4.csv']
news_content = pd.DataFrame()

for file in pyktok_files:
    pyktok_data = load_pyktok_data(f'/Users/catherinefoster/Desktop/CS315/cs315project2datacollectionV2/{file}')
    content = get_news_content(pyktok_data, news)
    news_content = pd.concat([content, news_content])
```
